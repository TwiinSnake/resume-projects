```python
import sqlite3
connection = sqlite3.connect("cse2050_movies_db.db")
cursor = connection.cursor()

sql = """
    CREATE TABLE imdb_movies (
    imdb_title_id VARCHAR(10),
    title VARCHAR(500),
    original_title VARCHAR(250),
    year INTEGER,
    date_published DATE,
    genre VARCHAR(200),
    duration INTEGER,
    country VARCHAR(100),
    language VARCHAR(50),
    director VARCHAR(250),
    writer VARCHAR(250),
    production_company VARCHAR(250),
    actors TEXT,
    description TEXT,
    avg_vote REAL,
    votes INTEGER,
    budget INTEGER,
    usa_gross_income INTEGER,
    worldwide_gross_income INTEGER,
    metascore INTEGER,
    reviews_from_users INTEGER,
    reviews_from_critics INTEGER,
    primary key(imdb_title_id)
    ) """

sql2 = """
    DROP TABLE IF EXISTS imdb_movies
    """

cursor.execute(sql2)
cursor.execute(sql)
connection.commit()
connection.close()
```

```python
import csv
import sqlite3
from datetime import datetime
import re

connection = sqlite3.connect("cse2050_movies_db.db")
cursor = connection.cursor()

with open("IMDb_movies.csv", encoding="utf-8") as file:
    reader = csv.reader(file, delimiter=',', quotechar='"')
    header = next(reader)
    for row in reader:
        # Remove beginning tt from ID
        row[0] = row[0][2:]
        # Convert date to datetime object
        if len(row[4]) < 5 :
            row[4] = datetime.strptime(row[4], "%Y")
        else :
            row[4] = datetime.strptime(row[4], "%Y-%m-%d")
        # Remove lettering from money related columns, as long as they are there.
        # Also, if they are there, convert from string to their respective data type.
        if row[16] :
            row[16] = re.sub('\D', '', row[16])
            row[16] = int(row[16])
        else :
            row[16] = None
        if row[17] :
            row[17] = re.sub('\D', '', row[17])
            row[17] = int(row[17])
        else :
            row[17] = None
        if row[18] :
            row[18] = re.sub('\D', '', row[18])
            row[18] = int(row[18])
        else :
            row[18] = None
        if row[19] :
            row[19] = int(float(row[19]))
        else :
            row[19] = None
        if row[20] :
            row[20] = int(float(row[20]))
        else :
            row[20] = None
        if row[21] :
            row[21] = int(float(row[21]))
        else :
            row[21] = None
        # Convert the other non-string rows to their respective data types.
        row[3] = int(row[3])
        row[6] = int(row[6])
        row[14] = float(row[14])
        row[15] = int(row[15])
        cursor.execute("INSERT INTO imdb_movies VALUES (?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?)", row)
        connection.commit()
connection.close()
```
